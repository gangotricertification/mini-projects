```python
import numpy as np
data=np.genfromtxt(r"H:\practice data\iris.data.txt",delimiter=",",dtype="str")
data
```

```python
X=data[:,[0,1,2,3]]
Y=data[:,[4]]
```

```python
print(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
setosa=mpimg.imread("H:\practice data\iris photo\setosa.jpg")
setosa
```

```python
versicolor=mpimg.imread("H:\\practice data\\iris photo\\versicolor.jpg")
versicolor
```

```python
virginica=mpimg.imread("H:\\practice data\\iris photo\\virginica.jpg")
print(virginica)
```

```python
d={'Iris-setosa':'setosa','Iris-versicolor':'versicolor','Iris-virginica':'virginica'}
```

```python
for i,y in enumerate(Y):
    Y[i][0]=d[y[0]]
Y
    

```

```python
from sklearn import linear_model
lr=linear_model.LogisticRegression()
lr.fit(X,Y)
```

```python
x=lr.predict([[2,2,2,1]])
x
```

```python
if x=='virginica':
    plt.imshow(virginica)
elif x=='setosa':
    plt.imshow(setosa)
else:
    plt.imshow(versicolor)
```
